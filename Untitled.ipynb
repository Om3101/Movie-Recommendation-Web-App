```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import mean_squared_error

# Load the CSV file
df = pd.read_csv('project.csv')
# Check the first few rows of the dataframe
print(df.head())

```

```python
# Load datasets
movies_df = pd.read_csv('project.csv')
ratings_df = pd.read_csv('ratings.csv')

# Display the first few rows
print("Movies Data:")
print(movies_df.head())

print("\nRatings Data:")
print(ratings_df.head())

# Check for missing values
print("\nMissing Values in Movies Data:")
print(movies_df.isnull().sum())

print("\nMissing Values in Ratings Data:")
print(ratings_df.isnull().sum())

# Shape of datasets
print("\nMovies Data Shape:", movies_df.shape)
print("Ratings Data Shape:", ratings_df.shape)
```

```python
# Distribution of genres
plt.figure(figsize=(10, 5))
movies_df['genre'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Genres')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()

# Overview length analysis
movies_df['overview_length'] = movies_df['overview'].fillna("").apply(len)
plt.figure(figsize=(10, 5))
sns.histplot(movies_df['overview_length'], bins=30, kde=True)
plt.title('Distribution of Overview Length')
plt.xlabel('Overview Length')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of ratings
plt.figure(figsize=(10, 5))
sns.histplot(ratings_df['rating'], bins=10, kde=False, color='orange')
plt.title('Ratings Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()

# Number of ratings per movie
plt.figure(figsize=(10, 5))
ratings_df['movie_id'].value_counts().head(10).plot(kind='bar', color='green')
plt.title('Top 10 Movies with Most Ratings')
plt.xlabel('Movie ID')
plt.ylabel('Number of Ratings')
plt.show()

# Number of ratings per user
plt.figure(figsize=(10, 5))
ratings_df['user_id'].value_counts().head(10).plot(kind='bar', color='purple')
plt.title('Top 10 Users with Most Ratings')
plt.xlabel('User ID')
plt.ylabel('Number of Ratings')
plt.show()
```

```python
# Pivot the ratings dataframe to create a user-item matrix
user_movie_matrix = ratings_df.pivot_table(index='user_id', columns='movie_id', values='rating').fillna(0)

# Inspect the matrix
print("User-Movie Matrix Shape:", user_movie_matrix.shape)
user_movie_matrix.head()
```

```python
train_df, test_df = train_test_split(ratings_df, test_size=0.2, random_state=42)

# Create user-item matrices for train and test sets
train_matrix = train_df.pivot_table(index='user_id', columns='movie_id', values='rating').fillna(0)
test_matrix = test_df.pivot_table(index='user_id', columns='movie_id', values='rating').fillna(0)

print("Train Matrix Shape:", train_matrix.shape)
print("Test Matrix Shape:", test_matrix.shape)
```

```python
# Initialize and train the KNN model
knn = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=5)
knn.fit(train_matrix)

print("KNN Model Trained Successfully!")
```

```python
# Ensure the test user ID exists in the training data
common_users = test_matrix.index.intersection(train_matrix.index)

if common_users.empty:
    raise KeyError("No common users found between train and test datasets.")

# Select a valid test user ID from common users
test_user_id = common_users[0]
print("Selected Test User ID:", test_user_id)

# Find nearest neighbors
distances, indices = knn.kneighbors(test_matrix.loc[test_user_id].values.reshape(1, -1), n_neighbors=5)
print("Distances:", distances)
print("Indices of Neighbors:", indices)

# Get similar users
similar_users = train_matrix.index[indices.flatten()]
print("Similar Users:", similar_users)

# Aggregate ratings from similar users
recommendations = train_matrix.loc[similar_users].mean(axis=0).sort_values(ascending=False)

# Get movies the test user has already rated
user_rated_movies = train_matrix.loc[test_user_id][train_matrix.loc[test_user_id] > 0].index

# Filter out movies the test user has already rated
recommendations = recommendations[~recommendations.index.isin(user_rated_movies)]

print("Top Recommended Movies:")
print(recommendations.head(10))
```

```python
# Calculate MSE between actual and predicted ratings
predicted_ratings = []
actual_ratings = []

for user_id in test_matrix.index:
    distances, indices = knn.kneighbors(test_matrix.loc[user_id].values.reshape(1, -1), n_neighbors=5)
    similar_users = train_matrix.index[indices.flatten()]
    recommendations = train_matrix.loc[similar_users].mean(axis=0)
    
    for movie_id in test_matrix.columns:
        if test_matrix.loc[user_id, movie_id] > 0:  # Only evaluate movies with ratings
            predicted_ratings.append(recommendations[movie_id] if movie_id in recommendations else 0)
            actual_ratings.append(test_matrix.loc[user_id, movie_id])

mse = mean_squared_error(actual_ratings, predicted_ratings)
print(f"Mean Squared Error (MSE): {mse}")
```

```python
import pickle

# Save the model
with open('knn_model.pkl', 'wb') as f:
    pickle.dump(knn, f)

print("KNN Model Saved!")
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# Define a threshold for "liking" a movie (e.g., rating >= 4)
LIKED_THRESHOLD = 4

# Predict movies for a test user
test_user_id = test_matrix.index[0]

# Get KNN recommendations for the test user
distances, indices = knn.kneighbors(test_matrix.loc[test_user_id].values.reshape(1, -1), n_neighbors=5)
similar_users = train_matrix.index[indices.flatten()]
recommendations = train_matrix.loc[similar_users].mean(axis=0).sort_values(ascending=False)

# Binarize the recommendations and actual ratings
recommended_movies = recommendations.index[:10]  # Top 10 recommendations
actual_likes = test_matrix.loc[test_user_id][test_matrix.loc[test_user_id] >= LIKED_THRESHOLD].index

# True Positive (TP): Recommended and liked
TP = len(set(recommended_movies).intersection(set(actual_likes)))

# False Positive (FP): Recommended but not liked
FP = len(set(recommended_movies) - set(actual_likes))

# False Negative (FN): Not recommended but liked
FN = len(set(actual_likes) - set(recommended_movies))

# True Negative (TN): Not recommended and not liked
total_movies = train_matrix.columns
not_recommended = set(total_movies) - set(recommended_movies)
not_liked = set(total_movies) - set(actual_likes)
TN = len(not_recommended.intersection(not_liked))

# Create the confusion matrix
conf_matrix = [[TP, FP], [FN, TN]]
print("Confusion Matrix:")
print(conf_matrix)

# Optional: Compute additional metrics
print("\nClassification Report:")
y_true = [1 if movie in actual_likes else 0 for movie in total_movies]
y_pred = [1 if movie in recommended_movies else 0 for movie in total_movies]
print(classification_report(y_true, y_pred))
```

```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

# Binarize actual and predicted ratings for all movies
y_true = [1 if movie in actual_likes else 0 for movie in total_movies]
y_scores = [recommendations.get(movie, 0) for movie in total_movies]

# Compute precision and recall
precision, recall, thresholds = precision_recall_curve(y_true, y_scores)

# Plot the Precision-Recall curve
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, marker='.', label="KNN")
plt.title('Precision-Recall Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc

# Compute False Positive Rate, True Positive Rate, and thresholds
fpr, tpr, thresholds = roc_curve(y_true, y_scores)
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.grid()
plt.show()
```

```python
import numpy as np

# Compute accuracy for top-N recommendations
top_n_values = [1, 5, 10, 20]
accuracies = []

for n in top_n_values:
    top_n_recommendations = recommendations.index[:n]
    correct_recommendations = set(top_n_recommendations).intersection(set(actual_likes))
    accuracies.append(len(correct_recommendations) / n)

# Plot the accuracy
plt.figure(figsize=(8, 6))
plt.bar(np.arange(len(top_n_values)), accuracies, color='skyblue')
plt.xticks(np.arange(len(top_n_values)), [f'Top-{n}' for n in top_n_values])
plt.title('Top-N Recommendation Accuracy')
plt.xlabel('Top-N Recommendations')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.grid(axis='y')
plt.show()
```
